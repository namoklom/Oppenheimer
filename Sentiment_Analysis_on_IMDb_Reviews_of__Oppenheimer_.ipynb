## **Step 2: Sentiment Analysis and Natural Language Processing (NLP)**


In this step, we perform sentiment analysis and natural language processing (NLP) on the text data.

First, we define a function called **analyze_sentiment**. This function takes a piece of text as input and utilizes the **TextBlob** library to analyze the sentiment of the text. **TextBlob** is a Python library that provides simple API for NLP tasks such as part-of-speech tagging, noun phrase extraction, sentiment analysis, and more.

The **analyze_sentiment** function calculates the sentiment polarity of the input text, which represents the degree of positivity or negativity of the sentiment expressed in the text. The polarity score ranges from -1 to 1, where a score closer to 1 indicates a more positive sentiment, a score closer to -1 indicates a more negative sentiment, and a score around 0 indicates a neutral sentiment.

After defining the **analyze_sentiment** function, we apply it to each text entry in the dataset's **'text'** column using the apply method. This allows us to analyze the sentiment of each text entry individually and store the sentiment polarity scores in a new column named **'sentiment'** within the **DataFrame**.



```python
def analyze_sentiment(text):
    '''
    Function to analyze the sentiment of text using TextBlob.
    Returns the polarity value from -1 to 1, where higher values indicate more positive sentiment.
    '''
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

df['sentiment'] = df['text'].apply(analyze_sentiment)
```

## **Step 3: Sentiment Analysis Result Evaluation**


In this step, we are evaluating the sentiment analysis results. First, we filter the **DataFrame (df)** to extract three subsets of reviews based on their sentiment polarity: positive reviews, negative reviews, and neutral reviews.

● **Positive reviews** are those with sentiment polarity values greater than 0.

● **Negative reviews** are those with sentiment polarity values less than 0.

● **Neutral reviews** are those with sentiment polarity values equal to 0.

We then calculate the total number of reviews in the dataset **(total_reviews)** as well as the counts of positive **(total_positive)**, negative **(total_negative)**, and neutral **(total_neutral)** reviews. These counts provide insights into the distribution of sentiment across the dataset.


```python
positive_reviews = df[df['sentiment'] > 0]
negative_reviews = df[df['sentiment'] < 0]
neutral_reviews = df[df['sentiment'] == 0]

total_reviews = len(df)
total_positive = len(positive_reviews)
total_negative = len(negative_reviews)
total_neutral = len(neutral_reviews)
```
